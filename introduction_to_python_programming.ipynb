## Numeric Data Types in Python




We will be focusing on two main numeric data for the time being:

**Integers**: Integers (or ints) are whole numbers without decimal points. In Python, integers can be positive, negative, or zero. They can be created by writing a number without a decimal point or by using the int() function.

**Floats**: Floats are numbers with decimal points. In Python, floats can be positive, negative, or zero. They can be created by writing a number with a decimal point, by using scientific notation (e.g. 1.23e-4), or by using the float() function.

**Booleans**: Boolean is a data type in Python that can have two values: True or False, used for logical operations.

Python also supports other numeric data types, such as complex numbers, bool for boolean values (True or False) and decimal for fixed-precision decimal values. Additionally, Python provides a rich set of operators and functions for performing arithmetic operations and other manipulations on numeric data.

### Best Practices/Tips to remember

1. **Write readable code:** Python is known for its readability, so it's important to write code that is easy to read and understand. Use meaningful variable names, add comments, and format your code consistently.
2. **Use code completion:** Most IDEs have a code completion feature that can save you time by automatically completing your code as you type. This can help you avoid syntax errors and make it easier to find the right function or method.
3. **Use debugging tools:** IDEs have built-in tools that allow you to step through your code, set breakpoints, and inspect variables. These tools can help you identify and fix bugs more quickly.
4. **Use version control:** Web-based development environments like Google Colab and Jupyter Notebook can integrate with version control systems like Git, which allows you to keep track of changes to your code and collaborate with others.
5. **Make use of Plugins and extensions:** IDEs usually have a lot of plugins and extensions that can help you to add new functionality to your development environment. Some of these plugins can help you to improve your productivity.
6. **Use PEP 8:** PEP 8 is a style guide for Python code that provides guidelines for writing readable and consistent code. Following these guidelines can make it easier for others to understand and maintain your code.

### **Shortcomings**

1. **Speed**: Python is an interpreted language, which can make it slower than compiled languages like C or C++.
2. **Memory management:** Python's memory management can be less efficient than languages like C or C++, which can lead to memory leaks and performance issues.
3. **Concurrency**: Python's Global Interpreter Lock (GIL) can make it challenging to take full advantage of multi-core processors.
4. **Error-prone**: Python's dynamic nature can make it more prone to errors, especially regarding type checking.
5. **Limited debugging capabilities:** Some IDEs for Python may not have advanced debugging features, such as the ability to step through code, set breakpoints, and inspect variables.

## <u>Arithmetic Precedence</u>

dffd.jpeg

In Python, like most programming languages, there is a set of rules that determine the order in which mathematical operations are performed, also known as "operator precedence." This means that certain operations are performed before others, which can affect the outcome of a calculation if operators with different precedence levels are used together.

BODMAS stands for Parentheses, Exponents, Multiplication and Division, and Addition and Subtraction. This means that when working with an expression, we should perform the following operations in this order:

1. Parentheses: Expressions inside parentheses are always evaluated first.
2. Exponents: The ** operator is evaluated next.
3. Multiplication and Division: *,  /,  // and % are evaluated from left to right next.
4. Addition and Subtraction: + and - are evaluated last, from left to right.

Here is a list of the operator precedence levels in Python, from highest to lowest:

Parentheses: Expressions inside parentheses are always evaluated first. For example:

```python
result = (2 + 3) * 4
# result is 20
```

Exponentiation: The ** operator is evaluated next. For example:

```python
result = 2 ** 3 * 4
# result is 32
```

Unary operators: +, and -, when used as unary operators, are evaluated next. For example:

```python
result = -2 ** 3 * 4
# result is -32
```

Multiplication and division: *, /, // and % are evaluated next. For example:

```python
result = 4 * 3 / 2
# result is 6.0
```

Addition and subtraction: + and - are evaluated last. For example:

```python
result = 2 + 3 - 4
# result is 1
```

If you use multiple operators with the same precedence level, they are evaluated from left to right. For example:

```python
result = 2 + 3 * 4
# result is 14
```

You can also use parentheses to change the order of evaluation. For example, in the above example, if you want to perform the addition operation first, you can use the following:

```python
result = (2 + 3) * 4
# result is 20
```

It is always a good practice to use parentheses when working with complex expressions to make the code more readable and prevent unexpected results due to operator precedence.
